```python
# How many people have heart disease and how many people  doesn't have heart disease?
df['target'].value_counts()
```

```python
df['target'].value_counts().plot(kind='bar', color=['orchid','salmon'])
plt.title('Heart Disease Values')
plt.xlabel('1 = Disease, 0 = No Disease ')
plt.ylabel('Amount')
plt.show()
```

```python
df.target.value_counts().plot(kind='pie',figsize=(8,6))
plt.title('Heart Disease Values')
plt.legend(['Disease', 'No Disease'])
plt.ylabel('target')
plt.show()
```

```python
df['sex'].value_counts()
```

```python
df['sex'].value_counts().plot(kind='pie',figsize=(8,6))
plt.legend(['Male','Female'])
plt.ylabel('sex')
plt.title('Haert Disease Gender Ratio');
```

```python
# People of which sex has most heart disease?
pd.crosstab(df.sex, df.target)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.ylabel('Amount');
```

```python

df.cp.value_counts()
```

```python
df['cp'].value_counts().plot(kind='bar', color=['purple','lightblue','lightgreen','yellow'])
plt.xlabel('Chest Pain Type')
plt.title('Chest pain vs count');
```

```python
#People of which sex has which type of chest pain most?
pd.crosstab(df.sex, df.cp)
```

```python
pd.crosstab(df.sex, df.cp).plot(kind='bar',color=['coral','khaki','lightgreen','yellow'])
plt.title("Types of Chest Pain w.r.t sex")
plt.xlabel('0 =Female , 1 = Male ');
```

```python
#  People with which chest pain are most pron to have heart disease?'

pd.crosstab(df.cp,df.target)
```

```python
sns.countplot(x='cp',data=df,hue='target')
plt.title(" Most Chest Pain to Heart Disease");
```
